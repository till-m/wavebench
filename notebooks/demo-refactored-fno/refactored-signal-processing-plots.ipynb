```python
import torch
import matplotlib.pyplot as plt
from wavebench.nn.fno import FNO2d
from wavebench.plot_utils import get_average_channel_variance_by_depth
from wavebench.plot_utils import get_average_channel_squared_mean_by_depth
from wavebench.plot_utils import plot_spp
```

```python

model = FNO2d(modes1=12,
      modes2=12,
      hidden_width=32,)

x = torch.rand(64, 1, 64, 64)
name_values = get_average_channel_squared_mean_by_depth(model, x)
fig = plt.figure(figsize=(8, 4))
plot_spp(name_values)
plt.show()



```

```python

name_values = get_average_channel_variance_by_depth(model, x)
fig = plt.figure(figsize=(8, 4))
plot_spp(name_values)
plt.show()
```
